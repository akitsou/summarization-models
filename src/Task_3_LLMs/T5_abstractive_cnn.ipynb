```python
import numpy as np
import pandas as pd
```

```python
training_df = pd.read_csv('/content/drive/MyDrive/NLU/Task_3/final_processed-cleaned_train.csv')
training_df
```

```python
MAX_LEN = 512
SUMMARY_LEN = 150
TRAINING_SIZE = 5000
```

```python
training_df = training_df.iloc[0:TRAINING_SIZE,:].copy()

trainning_article_ls = list(training_df['boiler-free_art'])
trainning_highlight_ls = list(training_df['boiler-free_high'])

del training_df
```

```python
from random import sample

for index in sample(list(np.arange(len(trainning_article_ls))),3):
    print('Original Text : ')
    print(trainning_article_ls[index])

    print('\n\nSummary Text : ')
    print(trainning_highlight_ls[index])
    print('===========================================================================================================\n\n')
```

```python
print(len(trainning_article_ls))
print(len(trainning_highlight_ls))
```

```python
df = pd.DataFrame(columns=['target_text','source_text'])
df['target_text'] = trainning_highlight_ls
df['source_text'] = ['summarize: '+item for item in trainning_article_ls]
```

```python
df.head()
```

```python
from simplet5 import SimpleT5

model = SimpleT5()
model.from_pretrained(model_type="t5", model_name="t5-base")
```

```python
MAX_EPOCHS = 5
```

```python
model.train(train_df=df[0:(int)(0.7*TRAINING_SIZE)],
            eval_df=df[(int)(0.7*TRAINING_SIZE):TRAINING_SIZE],
            source_max_token_len=MAX_LEN,
            target_max_token_len=SUMMARY_LEN,
            batch_size=8, max_epochs=MAX_EPOCHS, use_gpu=True)
```

```python
import os
model_path = ''
rootdir = 'outputs/'
for it in os.scandir(rootdir):
    if it.is_dir():
        if 'simplet5-epoch-'+(str)(MAX_EPOCHS-1) in it.path:
            model_path = it.path
            print(model_path)
```

```python
model.load_model("t5","./"+model_path, use_gpu=True)
```
