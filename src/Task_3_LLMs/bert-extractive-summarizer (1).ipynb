```python
df=pd.read_csv('/kaggle/input/boiler-free-test/final_processed-cleaned_test.csv')
df
```

```python
# Function to apply summarization using bert-extractive-summarizer
def generate_summary(text):
    summarizer = Summarizer()
    summary = summarizer(text, num_sentences=3)
    return summary

# Apply the function to each row in the 'article' column
df['summary'] = df['boiler-free_art'].apply(generate_summary)

# Write the summaries to a text file
with open('summaries.txt', 'w', encoding='utf-8') as file:
    for summary in df['summary']:
        file.write(summary + '\n')
```

```python
from rouge import Rouge
import pandas as pd

# Assuming 'df_selected' is your DataFrame with 'summary' and 'highlights' columns

def calculate_rouge2_scores(predicted, target):
    rouge = Rouge()
    scores = rouge.get_scores(predicted, target, avg=True)
    return scores['rouge-2']['p'], scores['rouge-2']['r'], scores['rouge-2']['f']

# Apply the ROUGE-2 calculation to each row
df['rouge2_precision'], df['rouge2_recall'], df['rouge2_fmeasure'] = zip(*df.apply(lambda row: calculate_rouge2_scores(row['summary'], row['boiler-free_high']), axis=1))

# Print or use the scores as needed
print("Average Rouge-2 Precision:", df['rouge2_precision'].mean())
print("Average Rouge-2 Recall:", df['rouge2_recall'].mean())
print("Average Rouge-2 F-measure:", df['rouge2_fmeasure'].mean())

# Calculate standard deviation
std_precision = df['rouge2_precision'].std()
std_recall = df['rouge2_recall'].std()
std_fmeasure = df['rouge2_fmeasure'].std()

print("Standard Deviation Rouge-2 Precision:", std_precision)
print("Standard Deviation Rouge-2 Recall:", std_recall)
print("Standard Deviation Rouge-2 F-measure:", std_fmeasure)
```
